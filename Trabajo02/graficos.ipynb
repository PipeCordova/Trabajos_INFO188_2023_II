```python
import matplotlib.pyplot as plt

## Gráfico de SpeedUp modo CPU 1 threads vs Modo CPU nt threads
# Sp = tiempo_cpu_1_thread / tiempo_cpu_nt_threads
# 2 <= nt <= 10 

# Eje x esta en escala lineal, por lo tanto la linea si es una linea recta

# Datos proporcionados
nt = [2, 3, 4, 5, 6, 7, 8, 9, 10]
speed_up = [1.73, 2.52, 3.34, 4.14, 5.0, 5.76, 6.49, 7.23, 8.01]

# Graficar nt vs Speed up
plt.plot(nt, speed_up, marker='o', linestyle='-')

# Etiquetas y título del gráfico
plt.xlabel('N° Threads')
plt.ylabel('Speed up')
plt.title('Speed Up Modo CPU: 1 Thread vs nt Threads')

# Mostrar etiquetas con valores encima de cada punto
for i, txt in enumerate(speed_up):
    plt.annotate(f'{txt:.2f}', (nt[i], txt + 0.1), ha='center', va='bottom')

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt

# Para este grafico fui cambiando el valor de n (exponencialmente) y fui comparando el tiempo que demora el
# modo CPU vs el modo GPU
# Sp = tiempo_cpu / tiempo gpu

# El eje x esta en escala logaritmica, por lo que la curva, en realidad es una linea recta

# Datos proporcionados
n = [2**10, 2**11, 2**12, 2**13, 2**14, 2**15, 2**16]
speed_up = [21.05, 63.08994709, 209.2694064, 863.5067873, 3224.668122, 9912.318792, 39491.65217]

# Graficar n vs Speed Up
plt.plot(n, speed_up, marker='o', linestyle='-', label='Speed Up')

# Etiquetas y título del gráfico
plt.xscale('log', base=2)  # Utiliza escala logarítmica en el eje x
plt.xlabel('n (Tamaño matriz)')
plt.ylabel('Speed Up')
plt.title('Speed Up - CPU (1 Thread) vs GPU')

# Mostrar etiquetas con valores redondeados encima de cada punto
for i, txt in enumerate(speed_up):
    plt.annotate(f'{round(txt)}', (n[i], txt + 200), ha='center', va='bottom')

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt

# Para este grafico fui aumentando el valor de n (exponencialmente), y fui registrando la memoria usada
# Utilizé densidad = 0.3, en teoría, nuestro método utiliza 1/3 de memoria de lo que usaría el método tradicional
# Si la densidad fuera 0.25, entonces se utilizaría 1/4 de la memoria que usaría el método tradicional
# El eje x esta en escala logaritmica, por lo que la curva, en realidad es una linea recta


# Datos proporcionados
n = [2**8, 2**9, 2**10, 2**11, 2**12, 2**13, 2**14]
nuestro_metodo_mb = [0.075, 0.301, 1.203, 4.803, 19.195, 76.808, 307.203]
metodo_tradicional_mb = [0.25, 1, 4, 16, 64, 256, 1024]

# Graficar las dos líneas
plt.plot(n, nuestro_metodo_mb, marker='o', linestyle='-', label='Nuestro Método')
plt.plot(n, metodo_tradicional_mb, marker='o', linestyle='-', label='Método Tradicional')

# Etiquetas y título del gráfico
plt.xscale('log', base=2)  # Utiliza escala logarítmica en el eje x
plt.xlabel('n (Tamaño de la matriz)')
plt.ylabel('Memoria (Mb)')
plt.title('Memoria Utilizada en Mb (d = 0.3)')

# Mostrar la leyenda
plt.legend()

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
